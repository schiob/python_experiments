```python
from __future__ import division
from numpy import *
 
class Tableau:
 
    def __init__(self, obj):
        self.obj = [1] + obj
        self.rows = []
        self.cons = []
 
    def add_constraint(self, expression, value):
        self.rows.append([0] + expression)
        self.cons.append(value)
 
    def _pivot_column(self):
        low = 0
        idx = 0
        for i in range(1, len(self.obj)-1):
            if self.obj[i] < low:
                low = self.obj[i]
                idx = i
        if idx == 0: return -1
        return idx
 
    def _pivot_row(self, col):
        rhs = [self.rows[i][-1] for i in range(len(self.rows))]
        lhs = [self.rows[i][col] for i in range(len(self.rows))]
        ratio = []
        for i in range(len(rhs)):
            if lhs[i] == 0:
                ratio.append(99999999 * abs(max(rhs)))
                continue
            ratio.append(rhs[i]/lhs[i])
        return argmin(ratio)
 
    def display(self):
        print '\n', matrix([self.obj] + self.rows)
 
    def _pivot(self, row, col):
        e = self.rows[row][col]
        self.rows[row] /= e
        for r in range(len(self.rows)):
            if r == row: continue
            self.rows[r] = self.rows[r] - self.rows[r][col]*self.rows[row]
        self.obj = self.obj - self.obj[col]*self.rows[row]
 
    def _check(self):
        if min(self.obj[1:-1]) >= 0: return 1
        return 0
         
    def solve(self):
 
        # build full tableau
        for i in range(len(self.rows)):
            self.obj += [0]
            ident = [0 for r in range(len(self.rows))]
            ident[i] = 1
            self.rows[i] += ident + [self.cons[i]]
            self.rows[i] = array(self.rows[i], dtype=float)
        self.obj = array(self.obj + [0], dtype=float)
 
        # solve
        self.display()
        while not self._check():
            c = self._pivot_column()
            r = self._pivot_row(c)
            self._pivot(r,c)
            print '\npivot column: %s\npivot row: %s'%(c+1,r+2)
            self.display()
"""             
if __name__ == '__main__':
 

    max z = 2x + 3y + 2z
    st
    2x + y + z <= 4
    x + 2y + z <= 7
    z          <= 5
    x,y,z >= 0
 
    t = Tableau([-2,-3,-2])
    t.add_constraint([2, 1, 1], 4)
    t.add_constraint([1, 2, 1], 7)
    t.add_constraint([0, 0, 1], 5)
    t.solve()

"""
```

```python
# 60 last index values
# 12 values per day
# 79-84 indexes of restrictions

lunes = list()
martes = list()
miercoles = list()
jueves = list()
viernes = list()

horas_por_maestro = list()

with open('horarios_finales.csv') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in islice(spamreader,3 , 58):
        lunes.append(row[:13] + ['1000' for _ in xrange(46)] + list(row[80]))
        martes.append([row[0],] + ['1000' for _ in xrange(1,13)] + row[13:25] + ['1000' for _ in xrange(25, 61)] + [row[81]])
        miercoles.append([row[0]] + ['1000' for _ in xrange(1,25)] + row[25:37] +['1000' for _ in xrange(37,61)] + [row[82]])
        jueves.append([row[0]] + ['1000' for _ in xrange(1,37)] + row[37:49] + ['1000' for _ in xrange(49,61)] + [row[83]])
        viernes.append([row[0]] + ['1000' for _ in xrange(1,49)] + row[49:61] + [row[84]])
        
        horas_por_maestro.append(row[:61] + list(row[79]))

for row in chain(['LUNES\n',], lunes, 
                 ['\nMARTES\n',], martes, 
                 ['\nMIERCOLES\n',], miercoles, 
                 ['\nJUEVES\n',], jueves, 
                 ['\nVIERNES\n',], viernes,
                 ['\nHORAS TOTALES POR MAESTRO\n',], horas_por_maestro):
    print(row)

print(len(lunes+martes+miercoles+jueves+viernes))
```
