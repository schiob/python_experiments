```python
import matplotlib.pyplot as plt
import pandas as pn
import numpy as np
import math
import seaborn
```

```python
xs = [23]
nex = (553 + 121 * xs[-1]) % 177
while nex != xs[0]:
    xs.append(nex)
    nex = (553 + 121 * nex) % 177

print("Periodo de vida:",xs, 'Tamaño del ciclo:', len(xs), sep='\n')
```

```python
df = pn.DataFrame()
df['Xi'] = xs
df['Ri'] = [x/176 for x in xs]
df
```

#### Prueba de Media

```python
r_media = sum(df['Ri'])/len(df['Ri'])
# alfa lo tomé como 0.05 por lo que Z alfa medios = 1.96
L_i_m = 1/2 - 1.96*(1/math.sqrt(12*len(df['Ri'])))
L_s_m = 1/2 + 1.96*(1/math.sqrt(12*len(df['Ri'])))
print('Media:', r_media, '\nLimite inferior:', L_i_m, '\nLímite superior:', L_s_m)
```

#### Prueba de Varianza

```python
var = sum(((x-r_media)**2 for x in df['Ri'])) / (len(df['Ri'])-1)
# alfa lo tomé como 0.05 por lo que X2 con 86 grados de livertad = 112 y X2 con -alfa = 62.3
L_i_v = 112 / (12*(len(df['Ri'])-1))
L_s_v = 62.3 / (12*(len(df['Ri'])-1))
print('Varianza:', var, '\nLimite inferior:', L_i_v, '\nLímite superior:', L_s_v)
```

#### Prueba de Uniformidad

```python
dic = {1/math.sqrt(87) * x : 0 for x in range(10)}
for r in df['Ri']:
    key = -1
    for k in dic:
        if r >= k > key:
            key = k
    dic[key] += 1
pu = pn.DataFrame()
pu['Intervalo'] = ['[{} - {})'.format(1/math.sqrt(87) * x, 1/math.sqrt(87) * (x+1)) for x in range(10)]
pu['Intervalo'][9] = '[0.9649012813540153 - 1.0)'
pu['Oi'] = [dic[1/math.sqrt(87) * x] for x in range(10)]
pu['Ei'] = 87 // math.sqrt(87)
pu['(Ei-Oi)²/Ei'] = ((pu['Ei']- pu['Oi'])**2) / pu['Ei']
pu.plot()
pu
```
