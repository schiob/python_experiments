```python
import pandas as pn
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
raw = pn.io.json.read_json('data/HistorialdeUbicaciones.json')
```

```python
raw.head()
```

```python
df = raw['locations'].apply(pn.Series)
```

```python
df.head()
```

```python
df['latitude'] = df['latitudeE7'] * 0.0000001
df['longitude'] = df['longitudeE7'] * 0.0000001
```

```python
fig = plt.figure(figsize=(20,10))

# Create a map, using the Gall–Peters projection, 
maps = Basemap(projection='gall',  
              # with low resolution,
              resolution = 'l', 
              # And threshold 100000
              area_thresh = 100000.0,
              # Centered at 0,0 (i.e null island)
              lat_0=0, lon_0=0,
              llcrnrlon=-119.33, llcrnrlat=15.11,
              urcrnrlon=-85.67, urcrnrlat=34.41)

# Draw the coastlines on the map
maps.drawcoastlines()

# Draw country borders on the map
maps.drawcountries()

# Fill the land with grey
maps.fillcontinents(color = '#888888')

# Draw the map boundaries
maps.drawmapboundary(fill_color='#f4f4f4')

# Define our longitude and latitude points
x,y = maps(df['longitude'].values, df['latitude'].values)

# Plot them using round markers of size 6
maps.plot(x, y, 'bo', markersize=6)

# Show the map
plt.show()  
```
