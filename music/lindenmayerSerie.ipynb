```python
axioma = raw_input('Original ')
```

#### Reglas

```python
rule = {'C4': 'C4 G4 ',
        'D4': 'E4 F4 G4 ',
        'E4': 'D4 E4 C4 ',
        'F4': 'G4 E4 B4 ',
        'G4': 'G4 A4 ',
        'A4': 'C4 D4 A4 ',
        'B4': 'C4 '}
```

```python
inter = 5

stream1 = stream.Stream()
for n in axioma.split():
    stream1.append(note.Note(n))
escalas = [stream1]

for x in range(inter):
    escalas.append(stream.Stream())
    temp = ''
    for n in axioma.split():
        temp += rule[n]
    for n in temp.split():
        escalas[x+1].append(note.Note(n))
    axioma = temp
    
    
```
