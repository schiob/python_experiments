```python
tonica = note.Note('A4')
print tonica.frequency, tonica.fullName
```

```python
razon = {1: 1.0,
         2: 9/8.0,
         3: 81/64.0,
         4: 4/3.0,
         5: 3/2.0,
         6: 27/16.0,
         7: 243/128.0,
         8: 2}

print 'Escala pitagórica---------------------------------'
escPit = list()
for x in xrange(1, 9):
    n = note.Note()
    n.frequency = tonica.frequency * razon[x]
    escPit.append(n)
    print n.frequency, '\t', n.fullName
```

```python
pit = stream.Part()
for n in escPit:
    pit.append(n)
pit.show()
```

```python
pit.show('midi')
```

```python
print 'Escala moderna---------------------------------' 
r = 2 ** (1/12.0)
escMod = list()
escMod.append(tonica)
aux = tonica
print n.frequency, '\t', n.fullName
for x in xrange(1, 13):
    n = note.Note()
    n.frequency = aux.frequency * (r)
    aux = n
    if not x in (1, 3, 6, 8, 10):
        escMod.append(n)
        print n.frequency, '\t', n.fullName
```

```python
mod = stream.Part()
for n in escMod:
    mod.append(n)
mod.show()
```

```python
mod.show('midi')
```

```python
dos = stream.Score()
dos.insert(0, pit)
dos.insert(0, mod)
dos.show()
```
